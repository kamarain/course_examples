# Introduction to OpenAI Gym Taxi Environment

## Load necessary libs and initialize the environment

```python
# Load OpenAI Gym and other necessary packages
import gym
import random
import numpy
import time

# Environment
env = gym.make("Taxi-v3")
env.reset()
env.render()
```

```python
## Run max 50 action steps and render output
```

```python
for t in range(50):
    action = input('Give action number: ')
    new_state, reward, done, info = env.step(int(action))
    print('New state: %d Reward: %d Done: %d Info: %s' %(new_state,reward,done,info))
    env.render()
```
